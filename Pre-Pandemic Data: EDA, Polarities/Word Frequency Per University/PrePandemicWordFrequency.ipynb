## Cornell Data

```python
#Load Cornell Data and Filter to necessary Columns
cornell = pd.read_csv('cornellprepandemic.csv')
def fillnaf(tbl):
    tbl1 = tbl["Body"].fillna(tbl["Title"])
    tbl["Body"] = tbl1
    return tbl
cornell = fillnaf(cornell)
cornell_data = cornell['Body']
cornell_data.index = cornell['Publish Date']

#Creating new columns for total counts/averages to be used later
afraid = [0]*cornell_data.shape[0]
anxiety = [0]*cornell_data.shape[0]
scared = [0]*cornell_data.shape[0]
lonely = [0]*cornell_data.shape[0]
stressed = [0]*cornell_data.shape[0]
helpw = [0]*cornell_data.shape[0]
failure =  [0]*cornell_data.shape[0]
death = [0]*cornell_data.shape[0]
depressed =  [0]*cornell_data.shape[0]
happy =  [0]*cornell_data.shape[0]
good = [0]*cornell_data.shape[0]
```

```python
#Code for accumulating the frequency of each word per month
for i in range(cornell_data.shape[0]):
    words = TextBlob(cornell_data[i]).words
    for word in words:
        if word == "afraid": afraid[i]=1
        if word == "anxiety" or word == 'anxious': anxiety[i]=1
        if word == "scared": scared[i]=1
        if word == "lonely": lonely[i]=1   # as all headlines are in lowercase
        if word == "stressed" or word == "stress": stressed[i]=1 
        if word == "help": helpw[i]=1  
        if word == 'fail' or word == 'failure': failure[i]= 1
        if word == 'death' or word == 'suicide' or word == 'suicidal': death[i] = 1
        if word == 'happy' : happy[i] = 1
        if word == 'good' : good[i] = 1
        if word == 'depressed' or word == 'depression' or word == 'depress': depressed[i]= 1
cornellwords = pd.DataFrame({'text':cornell_data,
                        'Afraid':afraid,
                        'Anxiety':anxiety,
                        'Scared':scared,
                        'Lonely':lonely,
                        'Stressed':stressed,
                        'Help': helpw,
                        'Failure': failure,
                        'Death': death,
                        'Happy': happy,
                        'Good': good,
                        'Depressed': depressed},
                        index=cornell_data.index)
cornellwords.head()
```

```python
#Accumulating the frequencies per month
cornellwords.index = pd.to_datetime(cornellwords.index)
#Samples monthly data for keywords
cornellmonthly = cornellwords.resample('M').sum()
cornellmonthly.drop(cornellmonthly.tail(1).index,inplace=True) #drop unnecessary last row
cornellmonthly['University'] = 'Cornell'

#Converting the results to a csv for Interactive Dashboard
cornellmonthly.to_csv('cornellpre.csv')
```
